¿De qué se tratan entonces las agrupaciones? Cuando agrupemos según una columna A, estaremos juntando todas las filas que tengan el mismo valor en dicha columna columna A, y luego aplicaremos una agregación sobre todos los valores de una columna B que hayan caído en cada grupo.
Si quisiéramos saber cuál es el valor total de infracciones de cada patente, deberíamos agrupar según la columna patente, colocando juntas todas las filas que tengan la misma patente y luego, por cada grupo, realizar la sumatoria de la columna infraccion.
